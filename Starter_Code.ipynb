## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns 'EIN' and 'NAME'

print(application_df.columns)

# Drop the non-beneficial ID columns 'EIN' and 'NAME'
if 'EIN' in application_df.columns and 'NAME' in application_df.columns:
    application_df = application_df.drop(columns=['EIN', 'NAME'])
else:
    print("Columns 'EIN' and 'NAME' not found in the DataFrame.")
```

```python
# Determine the number of unique values in each column.

unique_counts = application_df.nunique()

# Display the unique value counts for each column
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")

application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff = 150
application_types_to_replace = application_type_counts[application_type_counts < cutoff].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning

classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

classification_counts_gt1 = classification_counts[classification_counts > 1]
print(classification_counts_gt1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Choose a cutoff value
cutoff_value = 150  # You can adjust this value based on your preference

# Create a list of classifications to be replaced
classifications_to_replace = classification_counts[classification_counts < cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Use pd.get_dummies to convert categorical data to numeric
application_df = pd.get_dummies(application_df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'])

# Display the resulting DataFrame
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
# Define the target variable
y = application_df['IS_SUCCESSFUL']

# Define the features (all columns except 'IS_SUCCESSFUL')
X = application_df.drop(columns=['IS_SUCCESSFUL'])


# Split the preprocessed data into a training and testing dataset
from sklearn.model_selection import train_test_split

# Split the data into a training set (80%) and a testing set (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```

```python

# Select only numeric columns for scaling
numeric_columns = ['STATUS', 'ASK_AMT'] + [col for col in X.columns if col.startswith('APPLICATION_TYPE_') or col.startswith('AFFILIATION_') or col.startswith('CLASSIFICATION_') or col.startswith('USE_CASE_') or col.startswith('ORGANIZATION_') or col.startswith('INCOME_AMT_') or col.startswith('SPECIAL_CONSIDERATIONS_')]

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_train_scaled = scaler.fit_transform(X_train[numeric_columns])

# Scale the testing data using the same scaler
X_test_scaled = scaler.transform(X_test[numeric_columns])
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()
from tensorflow.keras.layers import Dense

# First hidden layer
input_features = X_train_scaled.shape[1]  # Number of columns in the scaled features
nn.add(Dense(units=64, activation='relu', input_dim=input_features))

# Second hidden layer
nn.add(Dense(units=32, activation='relu'))

# Output layer
nn.add(Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model

nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model

fit_model = nn.fit(X_train_scaled, y_train, epochs=150)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
